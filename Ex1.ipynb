```python
# Import các thư viện
from transformers import pipeline
import matplotlib.pyplot as plt
import pandas as pd

# Khởi tạo pipeline phân tích cảm xúc
sentiment_pipeline = pipeline("sentiment-analysis")
```

```python
# Câu mẫu
sample_sentence = "Tôi rất thích sử dụng Hugging Face cho các tác vụ xử lý ngôn ngữ tự nhiên!"

# Phân tích cảm xúc
result = sentiment_pipeline(sample_sentence)

# In kết quả
print(f"Câu: '{sample_sentence}'")
print(f"Kết quả phân tích cảm xúc: {result}")
```

```python
# Trích xuất nhãn và điểm số
sentiment_label = result[0]['label']
sentiment_score = result[0]['score']

# Tạo DataFrame và vẽ biểu đồ thanh
data = {'Nhãn': [sentiment_label], 'Điểm': [sentiment_score]}
df = pd.DataFrame(data)

plt.figure(figsize=(6, 4))
plt.bar(df['Nhãn'], df['Điểm'], color=['green' if sentiment_label == 'POSITIVE' else 'red'])
plt.ylim(0, 1)
plt.title('Kết quả phân tích cảm xúc')
plt.xlabel('Nhãn cảm xúc')
plt.ylabel('Điểm tin cậy')
plt.show()

print("\n--- Bài tập 1 Hoàn thành ---")
```
